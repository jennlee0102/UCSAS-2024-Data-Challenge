```python
# extracting the date portion after the '-' character in the 'Date' column

def process_date(row):
    if '-' in row['Date']:
        date_part = row['Date'].split('-')[1]
        # if there's space after '-', then remove the space
        if ' ' in date_part:
            return date_part.strip()
        else:
            return date_part
    else:
        return row['Date']

df['Date'] = df.apply(process_date, axis=1)
df.Date.unique()
```

```python
# check if the date process causes any NAs.
df.Date.isna().sum()
```

```python
# convert Date datatype to datetime
month_mapping = {
    'Jan': '01', 'Feb': '02', 'Mar': '03',
    'Apr': '04', 'May': '05', 'Jun': '06', 'June': '06',
    'Jul': '07', 'Aug': '08', 'Sep': '09',
    'Oct': '10', 'Nov': '11', 'Dec': '12'
}

# df['Date'] in the format 'DD Month YYYY' then convert the datatype from object to datetime 
df['Date'] = df['Date'].apply(lambda x: x.split(' ')[2] + '-' + month_mapping[x.split(' ')[1]] + '-' + x.split(' ')[0])
df['Date'] = pd.to_datetime(df['Date'], format='%Y-%m-%d', errors='coerce')
```

```python
# preprocess punctuation marks, remove accent, and modify Asian names

from unidecode import unidecode

def remove_accents(input_str):
    return unidecode(input_str)

def function(full_name, country):
    if country in ['CHN', 'KOR', 'SGP', 'SIN', 'HKG', 'TPE']:
        processed_name = full_name.replace('-', '')
        name_parts = processed_name.split()
        if len(name_parts) > 2:
            processed_name = name_parts[0] + name_parts[1] + ' ' + ''.join(name_parts[2:])

    else:
        processed_name = full_name.replace('-', ' ')

        # double space to one space
        processed_name = processed_name.replace('  ', ' ')
    
        # remove accents
        processed_name = remove_accents(processed_name)
    
    return processed_name

df['FullName'] = df.apply(lambda row: function(row['Name'], row['Country']), axis=1)
# move 'FullName' column to the first position
df = df[['FullName'] + [col for col in df.columns if col != 'FullName']]

# randomly check if the names in Singapore are converted correctly
df[df.Country == 'SIN']
```

```python
# choose the short name for the gymnasts with different names

def rename2(full_name):
    # split the full name into parts
    name_parts = full_name.split()

    # condition 1: If the name has 3 words, drop the middle word
    if len(name_parts) == 3:
        processed_name = f"{name_parts[0]} {name_parts[2]}"

    # condition 2: If the name has 4 words, keep the 1st and 3rd words
    elif len(name_parts) == 4:
        processed_name = f"{name_parts[0]} {name_parts[2]}"
    else:
        processed_name = full_name

    return processed_name

df['FinalName'] = df['FullName'].apply(rename2)

# move 'FullName' column to the first position
df = df[['FinalName'] + [col for col in df.columns if col != 'FinalName']]
df.head()
```

```python
# for each competition, if an athlete played in two or more different countries in df.Country, then change all the country data for that athlete to the last-played country based on the df['Date']

# sort by 'Date' within each 'FinalName' group
df = df.sort_values(by=['FinalName', 'Date'])

# create a dictionary to store the last-played country for each athlete
last_played_country = {}

# create a list to store athletes who played in different countries
different_countries_athletes = []

# for each athlete
for name, group in df.groupby('FinalName'):
    # check if the athlete played in different countries
    if group['Country'].nunique() > 1:
        different_countries_athletes.append(name)
        
    # update 'Country' with the last-played country based on 'Date'
    last_played_country[name] = group.iloc[-1]['Country']

# update 'Country' column based on the last-played country for each athlete
df['Country'] = df['FinalName'].map(last_played_country)

df.head()
```

```python
# filter rows where Apparatus is 'VT1' or 'VT2'
filter = df[df['Apparatus'].isin(['VT1', 'VT2'])]

# identify duplicate rows
duplicate = filter[filter.duplicated(subset=['Name', 'Country', 'Competition', 'Round'], keep=False)]

# # create df with the duplicate rows where Apparatus == 'VT1' or 'VT2'
# duplicates_df = filter[duplicate]

# initialize dropped_df to store the dropped rows
dropped_df = pd.DataFrame(columns=df.columns)

# iterate over each group of duplicates and drop the row with the smaller 'Score'
for _, group in duplicate.groupby(['Name', 'Country', 'Competition', 'Round']):
    index_to_drop = group['Score'].idxmin()

    # append the dropped row to dropped_df
    dropped_df = pd.concat([dropped_df, df.loc[[index_to_drop]]])

    # drop the row from the original DataFrame
    df = df.drop(index_to_drop)

# num of dropped rows
print(f"Number of dropped rows: {len(dropped_df)}")

dropped_df.info()
```
